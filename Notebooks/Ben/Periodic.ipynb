```python
import pandas as pd

from google.cloud import bigquery
from pathlib import Path

from params import *

def get_data_from_bq(
    gcp_project: str,
    query: str):
    client = bigquery.Client()
    query_job = client.query(query)
    result = query_job.result()
    df = result.to_dataframe()
    return df
```
