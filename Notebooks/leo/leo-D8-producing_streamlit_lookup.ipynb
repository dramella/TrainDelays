```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('../../data/EK_all.csv')
df
```

```python
unique_station = list(set(list(df['STATION_OR']) + list(df['STATION_DES'])))
station_df = pd.DataFrame(unique_station).rename(columns={0:'Station Name'})
```

```python
lookup = pd.read_csv('../../mapping_data/stanox-lookup-lc.csv')
merged = station_df.merge(lookup, on='Station Name')
```

```python
merged.to_csv('../../mapping_data/tmp.csv', index=False)
```

```python
merged_nonan = pd.read_csv('../../mapping_data/tmp.csv')
merged_nonan['Station Name'] = merged_nonan['Station Name'].apply(lambda x: x.title())
merged_nonan.drop_duplicates(subset='Station Name', inplace = True)
merged_nonan.sort_values(by='Station Name', inplace = True)
```

```python
merged_nonan.to_csv('../../mapping_data/lookup_for_streamlit.csv', index=False)
```
